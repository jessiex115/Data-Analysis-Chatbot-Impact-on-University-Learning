## **Step 2: Sentiment Analysis of Text Question**

**Purpose:**

Analyze student feedback to gauge sentiment towards the chatbot.

**Tasks:**

1. Text Preprocessing:

  Clean feedback by removing punctuation and stopwords.
2. Sentiment Analysis:
  
  Classify feedback as positive, negative, or neutral.
3. Keyword Extraction:

  Identify and visualize frequently mentioned terms.
4. Visualizations:

  Display sentiment distribution (pie chart) and keywords (word cloud).

**Insight:**

Sentiment analysis reveals student satisfaction, highlighting strengths and areas for improvement.

**Outcome:**

Actionable insights guide chatbot enhancements based on student feedback.

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
import seaborn as sns
import string

# Ensure necessary resources are downloaded
nltk.download('vader_lexicon')
nltk.download('stopwords')
from nltk.corpus import stopwords

# Load the dataset
file_path = '/content/AI_Chatbots_Students_Attitude_Dataset_EN.csv'
data = pd.read_csv(file_path, delimiter=';', encoding='latin1')

# Focus on Q10
feedback_column = 'Q10'

# Text Preprocessing
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    text = text.lower()  # Convert to lowercase
    text = ''.join([char for char in text if char not in string.punctuation])  # Remove punctuation
    tokens = text.split()  # Tokenize
    tokens = [word for word in tokens if word not in stop_words]  # Remove stopwords
    return ' '.join(tokens)

data['cleaned_feedback'] = data[feedback_column].astype(str).apply(preprocess_text)

# Sentiment Analysis
sia = SentimentIntensityAnalyzer()

def analyze_sentiment(text):
    score = sia.polarity_scores(text)
    if score['compound'] > 0.05:
        return 'Positive'
    elif score['compound'] < -0.05:
        return 'Negative'
    else:
        return 'Neutral'

data['sentiment'] = data['cleaned_feedback'].apply(analyze_sentiment)

# Function to categorize words based on their sentiment
def categorize_words_by_sentiment(words):
    positive_words = []
    neutral_words = []
    negative_words = []

    # Words to exclude from all groups
    exclude_words = ['ai', 'artificial', 'intelligence', 'chatgpt', 'chatbot', 'nan', 'learning', 'either', 'use', 'education', 'teachers', 'students']

    for word in words:
        # Skip excluded words
        if word in exclude_words:
            continue

        score = sia.polarity_scores(word)  # Get the sentiment score of the word
        if score['compound'] > 0.05:
            positive_words.append(word)
        elif score['compound'] < -0.05:
            negative_words.append(word)
        else:
            neutral_words.append(word)

    # Remove "overused" from neutral and add to negative
    if 'overused' in neutral_words:
        neutral_words.remove('overused')
        negative_words.append('overused')

    return positive_words, neutral_words, negative_words

# Tokenize the cleaned feedback and categorize the words
all_words = ' '.join(data['cleaned_feedback']).split()
positive_words, neutral_words, negative_words = categorize_words_by_sentiment(all_words)

# Display all categorized words
print("All Positive Words:")
print(positive_words)

print("\nAll Neutral Words:")
print(neutral_words)

print("\nAll Negative Words:")
print(negative_words)

# Keyword Extraction
vectorizer = CountVectorizer(max_features=50, stop_words='english')
vectorized_data = vectorizer.fit_transform(data['cleaned_feedback'])
keywords = vectorizer.get_feature_names_out()
keyword_counts = vectorized_data.toarray().sum(axis=0)

# Define the words to exclude from all analyses
exclude_words = ['ai', 'artificial', 'intelligence', 'chatgpt', 'chatbot', 'nan']

# Filter the word frequencies to exclude the unwanted words
filtered_word_freq_dict = {word: count for word, count in zip(keywords, keyword_counts) if word not in exclude_words}

# Visualize the filtered Keywords (Word Cloud)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(filtered_word_freq_dict)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Frequently Mentioned Keywords in Feedback (Q10)', fontsize=16)
plt.show()

# Sentiment Distribution Visualization
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='sentiment', hue='sentiment', palette='pastel', dodge=False, legend=False)
plt.title('Sentiment Distribution of Feedback (Q10)')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np

# Load the dataset
file_path = '/AI_Chatbots_Students_Attitude_Dataset_EN.csv'
data = pd.read_csv(file_path, delimiter=';', encoding='latin1')

# Likert scale columns for clustering (Q5-Q9)
likert_columns = ['Q6.1', 'Q6.2', 'Q6.3', 'Q6.4', 'Q6.5', 'Q6.6', 'Q6.7', 'Q6.8', 'Q7.1', 'Q7.2', 'Q7.3', 'Q7.4']

# Apply the Likert mapping (Strongly Agree = 1, etc.)
likert_mapping = {
    "Strongly Agree": 5,
    "Agree": 4,
    "Neutral": 3,
    "Disagree": 2,
    "Strongly Disagree": 1
}

# Map Likert responses to numerical values
for col in likert_columns:
    data[col] = data[col].map(likert_mapping)

# Step 1: Standardize the data (important for clustering)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[likert_columns])

# Step 2: K-Means Clustering with Elbow Method

# Calculate the Sum of Squared Distances for different numbers of clusters
inertia = []
for k in range(1, 11):  # Test from 1 to 10 clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.show()

# Based on the Elbow Method, choose the optimal K (e.g., 3 or 4)

# Step 3: Perform K-Means Clustering with the chosen number of clusters (e.g., K=3)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

# Add the cluster labels to the original dataset
data['Cluster'] = clusters

# Visualizing the clusters
plt.figure(figsize=(8, 6))
sns.countplot(x='Cluster', data=data, hue='Cluster', palette='pastel', legend=False)
plt.title('Cluster Distribution (K-Means)')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.show()

# Step 4: Hierarchical Clustering (Dendrogram)

# Perform hierarchical/agglomerative clustering using the linkage method
linked = linkage(scaled_data, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, labels=data['Cluster'].values, orientation='top', color_threshold=0)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Students')
plt.ylabel('Distance')
plt.show()

# Group by clusters and calculate the mean for each Likert scale question (Q5-Q9)
likert_columns = ['Q6.1', 'Q6.2', 'Q6.3', 'Q6.4', 'Q6.5', 'Q6.6', 'Q6.7', 'Q6.8', 'Q7.1', 'Q7.2', 'Q7.3', 'Q7.4']

# Group by 'Cluster' and calculate the mean for each Likert scale column
cluster_means = data.groupby('Cluster')[likert_columns].mean()

# Display the mean values for each cluster
print(cluster_means)

```
